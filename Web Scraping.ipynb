```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
```



#LHR CAMPUS

```python
url = 'https://lhr.nu.edu.pk/faculty/'
response = requests.get(url)
```

```python
faculty = bs(response.content,'html.parser')
facultyLahohre = faculty.findAll('div', class_='facultyCard')
```

```python
id=[]
name=[]
designation=[]
hec=[]
education=[]
email=[]
department=[]
extension=[]
img=[]
```

```python
def DepartmentName(key):
  dictionary = {
    'fsc':'Computing',
    'ee': 'Electrical Engineering',
    'cv': 'Civil Engineering',
    'fsm':'Management',
    'ss':'Science & Humanities'
  }
  return dictionary[key]
```

```python
for i in range(len(facultyLahohre)):
  facultyLhr=facultyLahohre[i]
  img.append('https://lhr.nu.edu.pk'+facultyLhr.find('img',class_='card-img-top')['src'])
  email.append(facultyLhr.find('p',class_='mb-0').text)
  name.append(facultyLhr.find('h5', class_='text-center').text)
  designation.append(facultyLhr.find('p',class_='small').text.split('\n')[0])

  if facultyLhr.find('p',class_='small').text.find('HEC') > -1:
    hec.append(True)
  else:
    hec.append(False)

  try:
    id.append(int(facultyLhr.find('a',class_='faculty-link')['href'].split('/')[-1]))
  except:
    id.append(-1)

  try:
    link='https://lhr.nu.edu.pk'+facultyLhr.find('a',class_='faculty-link')['href']
    resp=requests.get(link)
    facultyPage = bs(resp.content,'html.parser')
    try:
      education.append(facultyPage.find('div',class_='col-lg-8').find('ul').find('li').text.strip())
    except:
      education.append('Not Found')

    try:
      ext=facultyPage.find('span',class_='small').text
      extension.append(int(ext[ext.find('Ext')+4:]))
    except:
      ext=0
      extension.append(ext)
  except:
    education.append('Not Found')
    extension.append(0)

  department.append(DepartmentName(facultyLhr.findParent('div',class_='container').findParent('div',class_='container')['id']))

print(len(id),len(name),len(designation),len(email),len(img),len(hec),len(education),len(extension),len(department))

```

```python
LahoreCampusData = pd.DataFrame({'ID':id,
                  'Name':name,
                  'Designation':designation,
                  'HEC Approved Approved PHD Supervisor':hec,
                  'Highest Education':education,
                  'Email':email,
                  'Department':department,
                  'Extension':extension,
                  'Image':img
                   })
```

```python
LahoreCampusData.to_csv('Lhr.csv', index=False)
```

#ISB CAMPUS

```python
HTMLFile = open("FAST NUCES Islamabad.html", "r")
index = HTMLFile.read()
```

```python
def DepartmentName(key):
  dictionary = {
    'CS':'Computer Science',
    'EE':'Electrical Engineering',
    'SH':'Science and Humanities',
    'MS':'Management Sciences',
  }
  return dictionary[key]
```

```python
id=[]
name=[]
designation=[]
hec=[]
education=[]
email=[]
department=[]
extension=[]
image=[]
```

```python
faculty = bs(index,'html.parser')
dataIsb = faculty.findAll('div', class_='col-md-12')

for i in range(len(dataIsb)-2):
  facultyDept=dataIsb[i].find('div',class_='row').findAll('div',class_='our-team')
  for j in range(len(facultyDept)):
    id.append(facultyDept[j].find('ul',class_='social').li.a['href'][-4:])
    name.append(facultyDept[j].find('h3',class_='title').text.strip())
    designation.append(facultyDept[j].find('span',class_='post').text.strip())
    link=facultyDept[j].find('a')['href']
    response = requests.get(link)
    facultyDetails = bs(response.content,'html.parser')
    try:
      if facultyDetails.find('div',class_='profile-info').findAll('p')[1].label.text.find('HEC') > -1:
        hec.append(True)
      else:
        hec.append(False)
    except:
      hec.append(False)
    
    

    options = Options()
    options.headless = True 
    options.add_argument("--window-size=1920,1200")
    driver = webdriver.Chrome(options=options)
    driver.get(link)
    try:
      edu = driver.find_element(By.ID,"DegreeName")
      education.append(edu.text.split('\n')[0])
      print(edu.text.split('\n')[0])
    except:
      education.append('Not Found')
    driver.quit()

    email.append(facultyDept[j].find('p').text)
    department.append(DepartmentName(dataIsb[i].find('div',class_='row')['id'][-2:]))
    try:
      temp=facultyDetails.find('div',class_='profile-info').findAll('p')[-1].span.text
      extension.append( int(temp[temp.find('Ext')+5:]))
    except:
      extension.append(0)
    image.append('http://isb.nu.edu.pk'+facultyDetails.find('div',class_='img-thumbnail').img['src'])



```

```python
print(len(id),len(name),len(designation),len(email),len(image),len(hec),len(education),len(extension),len(department))
```

```python
IslamabadCampusData = pd.DataFrame({'ID':id,
                  'Name':name,
                  'Designation':designation,
                  'HEC Approved Approved PHD Supervisor':hec,
                  'Highest Education':education,
                  'Email':email,
                  'Department':department,
                  'Extension':extension,
                  'Image':image
                   })
```

```python
IslamabadCampusData.to_csv('Isb.csv', index=False)
```

#KHI CAMPUS

```python
url = 'https://khi.nu.edu.pk/'
response = requests.get(url)

data = bs(response.content,'html.parser')
```

```python
id=[]
name=[]
designation=[]
hec=[]
education=[]
email=[]
department=[]
extension=[]
image=[]
```

```python
def DepartmentName(key):
  dictionary = {
    1:'Computer Science',
    2:'Cyber Security',
    3:'Artificial Intelligence',
    4:'Software Engineering',
    5:"Electrical ENgineering",
    6:"Management Sciences",
    7:"Science and Humanities"
  }
  return dictionary[key]
```

```python
departmentLinks = data.findAll('ul',class_='sub-menu')[1].findAll('a')
for i in range(1,8):
  deptLink = departmentLinks[i]['href']
  response = requests.get(deptLink)
  deptData = bs(response.content,'html.parser')
  id_=deptData.find("script",type="text/javascript",id="burst-js-extra").text
  index=id_.find("page_id")
  id_=int(id_[index+10:index+14])
  faculty = deptData.findAll('div',class_="gdlr-core-personnel-list-column")
  for j in range(len(faculty)):
    id.append(id_)
    try:
      image.append(faculty[j].find('img')["src"])
    except:
      image.append("Not Found")
    name.append(faculty[j].find('h3').text.split(',')[0].strip())
    email.append(faculty[j].find('div',class_="kingster-type-email").text)
    try:
      extension.append(int(faculty[j].find('div',class_="kingster-type-phone").text))
    except:
      extension.append(0)
    designation.append(faculty[j].find('div',class_="gdlr-core-personnel-list-position").text)
    link=faculty[j].find('a')['href']
    facultyDetails=(requests.get(link))
    facultyDetails = bs(facultyDetails.content,'html.parser')
    try:
      if facultyDetails.find('span',class_="gdlr-core-title-item-caption").text.find('HEC') > -1:
        hec.append(True)
      else:
        hec.append(False)
    except:
      hec.append(False)
    try:
      education.append(facultyDetails.find('span',class_="gdlr-core-icon-list-content").text)
    except:
      education.append('Not Found')
    department.append(DepartmentName(i))
```

```python
print(len(id),len(name),len(designation),len(email),len(image),len(hec),len(education),len(extension),len(department))
```

```python
KarachiCampusData = pd.DataFrame({'ID':id,
                  'Name':name,
                  'Designation':designation,
                  'HEC Approved Approved PHD Supervisor':hec,
                  'Highest Education':education,
                  'Email':email,
                  'Department':department,
                  'Extension':extension,
                  'Image':image
                   })
```

```python
KarachiCampusData.to_csv('Khi.csv', index=False)
```
